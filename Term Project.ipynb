```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
drama=pd.read_csv('kdrama.csv',sep=',')
drama.head()
```

```python
over_2017 = drama['Year of release'] > 2016
drama_over_2017 = drama[over_2017]
drama_rating=drama_over_2017.groupby('Year of release' )['Rating'].mean()
drama_rating.plot(kind='bar',ylim=(7,9),grid=True,ylabel="Rating average",color='red')
```

```python
genre_list = []
for gr in drama['Genre'].to_list():
    gr = gr.strip().split(", ")
    for a in gr:
        genre_list.append(a)
        
genre_df = pd.DataFrame.from_dict(Counter(genre_list),orient='index').rename(columns={0:'Count'})
genre_df.sort_values(by='Count',ascending = False,inplace = True)
genre_df 
```

```python
genre_delete=genre_df[genre_df['Count']<20].index
genre_most=genre_df.drop(genre_delete)
genre_most.plot(kind='bar',grid=True,ylabel="Count",xlabel="Genre")
```

```python
is_year1 = drama['Year of release'] == 2017
df_2017 = drama[is_year1]

genre_list = []
for gr in df_2017['Genre'].to_list():
    gr = gr.strip().split(", ")
    for a in gr:
        genre_list.append(a)
        
genre_2017_df = pd.DataFrame.from_dict(Counter(genre_list),orient='index').rename(columns={0:'Count'})
genre_2017_df.sort_values(by='Count',ascending = False,inplace = True)
genre_2017_most=genre_2017_df.loc[[' Drama', ' Romance', ' Comedy', ' Mystery',' Family',' Crime']]
genre_2017_most
```

```python
is_year2 = drama['Year of release'] == 2018
df_2018 = drama[is_year2]

genre_list = []
for gr in df_2018['Genre'].to_list():
    gr = gr.strip().split(", ")
    for a in gr:
        genre_list.append(a)
        
genre_2018_df = pd.DataFrame.from_dict(Counter(genre_list),orient='index').rename(columns={0:'Count'})
genre_df.sort_values(by='Count',ascending = False,inplace = True)
genre_2018_most=genre_2018_df.loc[[' Drama', ' Romance', ' Comedy', ' Mystery',' Family',' Crime']]
genre_2018_most
```

```python
is_year3 = drama['Year of release'] == 2019
df_2019 = drama[is_year3]

genre_list = []
for gr in df_2019['Genre'].to_list():
    gr = gr.strip().split(", ")
    for a in gr:
        genre_list.append(a)
        
genre_2019_df = pd.DataFrame.from_dict(Counter(genre_list),orient='index').rename(columns={0:'Count'})
genre_2019_df.sort_values(by='Count',ascending = False,inplace = True)
genre_2019_most=genre_2019_df.loc[[' Drama', ' Romance', ' Comedy', ' Mystery',' Family',' Crime']]
genre_2019_most
```

```python
is_year4 = drama['Year of release'] == 2020
df_2020 = drama[is_year4]

genre_list = []
for gr in df_2020['Genre'].to_list():
    gr = gr.strip().split(", ")
    for a in gr:
        genre_list.append(a)
        
genre_2020_df = pd.DataFrame.from_dict(Counter(genre_list),orient='index').rename(columns={0:'Count'})
genre_2020_df.sort_values(by='Count',ascending = False,inplace = True)
genre_2020_most=genre_2020_df.loc[[' Drama', ' Romance', ' Comedy', ' Mystery',' Family',' Crime']]
genre_2020_most
```

```python
genre_all_year=pd.concat([genre_2017_most,genre_2018_most,genre_2019_most,genre_2020_most],axis=1).set_axis(['2017','2018','2019','2020'],axis=1)

ys_drama=genre_all_year.loc[' Drama'].to_list()
ys_comedy=genre_all_year.loc[' Comedy'].to_list()
ys_romance=genre_all_year.loc[' Romance'].to_list()
ys_mystery=genre_all_year.loc[' Mystery'].to_list()
ys_family=genre_all_year.loc[' Family'].to_list()
ys_crime=genre_all_year.loc[' Crime'].to_list()
xs=genre_all_year.columns.to_list()

plt.figure(figsize=(10, 8))
plt.subplot(1,1,1)
plt.plot(xs,ys_drama,'o-',ms=3,lw=1,label='Drama')
plt.plot(xs,ys_comedy,'o-',ms=3,lw=1,label='Comedy')
plt.plot(xs,ys_romance,'o-',ms=3,lw=1,label='Romance')
plt.plot(xs,ys_mystery,'o-',ms=3,lw=1,label='Mystery')
plt.plot(xs,ys_family,'o-',ms=3,lw=1,label='Family')
plt.plot(xs,ys_crime,'o-',ms=3,lw=1,label='Crime')
plt.ylim(0,20)
plt.xlabel('Year of release')
plt.ylabel('Count')
plt.legend()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

model=LinearRegression()

x_train = np.reshape(xs,(-1,1))
y_train = np.reshape(ys_drama,(-1,1))
model.fit(x_train, y_train)
x_test = [[2022]]
y_pred = model.predict(x_test)

print(f'{x_test[0][0]} 년 Drama장르로 만드는 드라마의 예상 수는 {int(y_pred)} 입니다.')
```
